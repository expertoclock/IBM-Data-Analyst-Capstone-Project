# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, edgecolor='black')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Query to fetch Age data
QUERY = """
SELECT Age
FROM master
WHERE Age IS NOT NULL  -- Ensure Age column has no NULL values
"""

# Fetching data from SQLite database
df = pd.read_sql_query(QUERY, conn)

# Plotting the box plot
plt.figure(figsize=(8, 5))
plt.boxplot(df['Age'], vert=False)  # Horizontal box plot
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5)
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT LanguageWorkedWith, COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith IS NOT NULL
GROUP BY LanguageWorkedWith
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(8, 8))
plt.pie(df['count'], labels=df['LanguageWorkedWith'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Top 5 Languages/Technologies Respondents Worked with')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Query to fetch the necessary data
QUERY = """
SELECT Age,
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs,
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM master
WHERE Age >= 30 AND Age <= 35
GROUP BY Age
"""

# Fetch data from SQLite and calculate averages
df = pd.read_sql_query(QUERY, conn)

# Plotting the stacked bar chart
df.plot(x='Age', kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Query to fetch the necessary data
QUERY = """
SELECT Age,
       AVG(ConvertedComp) AS MedianConvertedComp
FROM (
    SELECT Age, ConvertedComp,
           ROW_NUMBER() OVER (PARTITION BY Age ORDER BY ConvertedComp) as RowNum,
           COUNT(*) OVER (PARTITION BY Age) as TotalRows
    FROM master
    WHERE Age >= 45 AND Age <= 60
)
WHERE RowNum IN (CEIL(TotalRows/2.0), FLOOR(TotalRows/2.0) + 1)
GROUP BY Age
"""

# Fetch data from SQLite and calculate averages
df = pd.read_sql_query(QUERY, conn)

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['MedianConvertedComp'], marker='o')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Line Chart of Median Converted Compensation (Age 45-60)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# SQL query to fetch MainBranch data and their counts
QUERY = """
SELECT MainBranch, COUNT(*) AS count
FROM master
GROUP BY MainBranch
"""

# Fetching data from SQLite database
df = pd.read_sql_query(QUERY, conn)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
colors = plt.cm.tab20c(range(len(df)))  # Color map for bars
plt.barh(df['MainBranch'], df['count'], color=colors)
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Horizontal Bar Chart of Main Branch')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest count at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add grid lines for better readability
plt.tight_layout()  # Ensure labels fit within the figure area
plt.show()
```
